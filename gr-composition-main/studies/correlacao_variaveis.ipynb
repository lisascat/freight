```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

## Utils

```python
def calcular_estatisticas(df, coluna):
    """
    Calcula e imprime as medidas de tendência central e dispersão para cada coluna de um DataFrame do pandas.

    Parâmetros:
    df (pandas.DataFrame): DataFrame para o qual as estatísticas serão calculadas.
    """
    # Medidas de tendência central
    media = df[coluna].mean()
    mediana = df[coluna].median()
    modas = df[coluna].mode()

    # Medidas de dispersão
    desvio_padrao = df[coluna].std(ddof=1)  
    variancia = df[coluna].var(ddof=1)      
    amplitude = df[coluna].max() - df[coluna].min()

    print(f"\nEstatísticas para '{coluna}':")
    print(f"- Média: {round(media,2)}")
    print(f"- Mediana: {round(mediana,2)}")
    print(f"- Moda(s): {', '.join(map(str, modas))}")
    print(f"- Desvio Padrão (para amostra): {round(desvio_padrao,2)}")
    print(f"- Variância (para amostra): {round(variancia,2)}")
    print(f"- Amplitude: {amplitude}")
```

```python
def plot_qq(data):
    """
    Cria um gráfico de quantis normais (Q-Q Plot) para um conjunto de dados.
    Este gráfico permite comparar seus dados com uma distribuição normal teórica. Se os dados seguirem uma linha reta no gráfico, isso é uma indicação de normalidade.

    Parâmetros:
    data (array-like): Conjunto de dados para o qual o gráfico será criado.
    """
    # Gerando o gráfico Q-Q
    plt.figure(figsize=(8, 6))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title("Gráfico de Quantis Normais (Q-Q Plot)")
    plt.ylabel("Quantis da Amostra")
    plt.xlabel("Quantis Teóricos da Distribuição Normal")

    plt.show()
```

```python
from scipy.stats import pointbiserialr, kendalltau
#homocedasticidade

def calculate_correlations(df, binary_vars, continuous_vars):
    correlations = {}

    for var in binary_vars:
        df[var] = df[var].astype(int)
    # Correlação entre variáveis binárias - Correlação de Phi (Pearson)
    for var1 in binary_vars:
        for var2 in binary_vars:
            if var1 != var2:
                corr_matrix = np.corrcoef(df[var1].dropna(), df[var2].dropna())
                corr = corr_matrix[0, 1]
                correlations[(var1, var2)] = corr

    # Correlação entre variáveis binárias e contínuas - Correlação de Ponto Bisserial
    for binary_var in binary_vars:
        for continuous_var in continuous_vars:
            corr, _ = pointbiserialr(df[binary_var].dropna(), df[continuous_var].dropna())
            correlations[(binary_var, continuous_var)] = corr
            correlations[(continuous_var, binary_var)] = corr

    # Correlação entre variáveis contínuas - Correlação de Kendall
    for var1 in continuous_vars:
        for var2 in continuous_vars:
            if var1 != var2:
                corr, _ = kendalltau(df[var1].dropna(), df[var2].dropna())
                correlations[(var1, var2)] = corr

    return pd.DataFrame.from_dict(correlations, orient='index', columns=['Correlation'])
```

```python
def plot_correlation_heatmap(correlation_df):
    """
    Plota um heatmap a partir de um DataFrame de correlações.

    :param correlation_df: DataFrame de correlações, onde o índice é uma tupla de pares de variáveis e 
                           as colunas contêm os valores de correlação.
    """

    correlation_df.index = pd.MultiIndex.from_tuples(correlation_df.index)
    correlation_df = correlation_df.round(2)
    
    correlation_matrix = correlation_df.unstack().droplevel(0, axis=1)

    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title("Correlation Heatmap")
    plt.show()
```
